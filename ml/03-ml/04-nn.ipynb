## Key Concepts

- **Activation Functions**: decide whether a neuron should be activated or not by calculating a weighted sum and adding bias. The purpose is to introduce non-linearity into the output of a neuron.
- **Weights**: parameters that transform input data within the network.
- **Bias**: threshold value that helps the model make better predictions.
- **Neoron**: basic unit of a neural network that processes input data and characterized by weights, bias, and activation function.
- **Input Layer**: the first layer of a neural network that receives input data.
- **Hidden Layers**: layers between input and output layers where computations are performed. Hidden layers often share the same activation function, while the output layer typically uses a different one depending on the model's prediction goal.
- **Output Layer**: the final layer that produces the output of the neural network.
- **Feedforward Propagation**: the process of moving input data through the network to generate an output. Input is used to calculate some intermediate function in the hidden layer, which is then used to calculate an output. Activation Function is a mathematical 'gate" in between the input feeding the current neuron and its output going to the next layer.
  - take input and multiply by neuron weights -> add bias -> apply activation function -> output to next layer.
- **Backpropagation**: the process of updating weights and biases based on the error of the output compared to the expected result. Aims to minimize the cost function by adjusting the network's weights and biases. The cost function gradients determine the level of adjustment with respect to parameters like activation function, weights, bias, etc.
  - calculate error -> compute gradients -> update weights and biases using optimization algorithm (e.g., gradient descent).

## Activation Function Types

- **Binary Step**: activates a neuron based on a threshold: if the input exceeds the threshold, the neuron activates; otherwise, it stays inactive and doesn't pass its output to the next layer. $f(x) = \begin{cases} 1 & \text{if } x \geq 0 \\ 0 & \text{if } x < 0 \end{cases} $
  - limitations:
    - It cannot provide multi-value outputs, e.g., it cannot be used for multi-class classification problems
    - gradient of the step function is zero, which causes a hindrance in the backpropagation process that will not allow the weights to be updated.
- **Linear Activation (identity)**: outputs the input directly without modification, making the activation proportional to the input. $f(x) = x$
  - limitations:
    - Backpropagation can't be used because the function's derivative is constant and unrelated to the input
      `x`
    - linear activation makes all layers collapse into one, turning the network into a single linear layer regardless of depth
- **Non-Linear Activation**: it captures complex input-output relationships by enabling backpropagation and allowing multiple layers to form non-linear combinations of inputs.
  - **Sigmoid/Logistic**: takes any real value as input and outputs values in the range of 0 to 1. As the input increases, the output approaches 1.0; as it decreases, the output approaches 0.0. Often used in binary classification problems. $f(x) = \frac{1}{1 + e^{-x}}$
    - pros:
      - used in models predicting probabilities (range 0-1), making sigmoid ideal due to its output range
      - differentiable with a smooth, S-shaped gradient that prevents abrupt output changes
    - cons:
      - prone to vanishing gradient problem for extreme input values, slowing down learning in deep networks
      - outputs not zero-centered, which can affect convergence during optimization
  - **Tanh**: outputs values between -1 and 1, often used in hidden layers.
  - **ReLU (Rectified Linear Unit)**: outputs the input directly if positive; otherwise, it outputs zero. Commonly used in hidden layers due to its efficiency.
  - **Leaky ReLU**: similar to ReLU but allows a small gradient when the unit is not active, helping to avoid the "dying ReLU" problem.
  - **Softmax**: used in the output layer for multi-class classification problems, converting logits into probabilities.

## Recommendations

GELU or Swish/SiLU for Transformers and most modern architectures; ReLU or Leaky ReLU for resource-constrained or very deep CNNs.

- **Dying ReLU** is a real production issue - monitor dead neurons during training.
- **Zero-centered** functions (Tanh, approximate GELU/ELU) often lead to faster convergence in very deep nets.
- **Self-normalizing** (SELU) can eliminate need for BatchNorm in some feed-forward setups, reducing memory.
- **Compute cost** matters at scale - ReLU family dominates edge/mobile; GELU/Swish acceptable on GPUs/TPUs.


# Gradient Descent

**Gradient Descent**: Iterative first-order optimization algorithm to minimize a differentiable loss function J(θ) by updating parameters θ in the direction opposite to the gradient. Core idea: θ ← θ - η ∇J(θ), where η is learning rate. Used universally in training neural networks, linear/logistic regression, and most ML models. Goal is to find local (or global in convex cases) minimum of the objective.

## Mathematical Formulation

- Gradient: $ \nabla J(\theta) = \left[ \frac{\partial J}{\partial \theta_1}, \dots, \frac{\partial J}{\partial \theta_n} \right] $
- Update rule: $ \theta\_{t+1} = \theta_t - \eta \nabla J(\theta_t) $
- For convex functions: Guaranteed convergence to global minimum with proper η
- For non-convex (deep learning): Converges to local minimum or saddle point

## Types of Gradient Descent

| Method                              | Description                                           | Advantages                                                                                                  | Disadvantages                                                                                 |
| ----------------------------------- | ----------------------------------------------------- | ----------------------------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------- |
| Batch Gradient Descent (Vanilla GD) | Computes gradient over entire dataset                 | Exact, Stable convergence, smooth trajectory                                                                | Computationally expensive for large datasets, Memory-intensive (O(N) where N is dataset size) |
| Stochastic Gradient Descent (SGD)   | Updates using gradient of single random example       | Noisy updates, faster iterations, escapes local minima better, Lower memory (O(1)), enables online learning | High variance, requires learning rate decay                                                   |
| Mini-Batch Gradient Descent         | Compromise: gradient over small batch (32-256 common) | Balances stability and speed, GPU-friendly, Default in deep learning frameworks (PyTorch, TensorFlow)       | Requires tuning batch size; some variance remains                                             |

## Learning Rate and Scheduling

- Critical hyperparameter: too high → divergence; too low → slow convergence
- Fixed: Simple but rarely optimal
- Schedules:
  - Step decay: Reduce η by factor every k epochs
  - Exponential decay: $ \eta_t = \eta_0 \gamma^t $
  - Cosine annealing: Smooth decay with restarts (SGDR)
  - Cyclical learning rates: Triangle policy for better exploration
  - 1-cycle policy: Single cycle with momentum adjustment
- Adaptive methods (see variants) often reduce need for manual scheduling

## Recommendations

Start with **AdamW** + cosine annealing or linear warmup + decay schedule. It delivers the best generalization in most supervised tasks.

- **When to switch to SGD + Momentum**: Large-scale vision models (e.g., ConvNeXt, EfficientNet) or when extreme fine-tuning for accuracy is needed - SGD often achieves lower final loss given sufficient tuning.
- **Learning rate schedules** (almost mandatory in production): Warmup (linear for 5-10% of steps) → cosine decay or exponential decay. One-cycle policy also popular.
- **Batch size impact**: Larger batches (with LR scaling, e.g., linear rule) often improve stability but may hurt generalization; 2048-8192 common with gradient accumulation in enterprise.
- **Memory-efficient variants**: Use AdamW-8bit or Lion (evo-norm based, less memory) for very large models on limited hardware.


# Forward vs Backward Propagation

| Aspect                           | Forward Propagation                                                                                                        | Backpropagation                                                                                                                                                                                                      |
| -------------------------------- | -------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Definition**                   | The process of passing input data through the network layers sequentially to compute the output prediction                 | The process of propagating the error gradients backward through the network to compute gradients of the loss with respect to weights                                                                                 |
| **Primary Purpose**              | Compute the network's output (prediction) and calculate the loss/error for a given input                                   | Compute gradients of the loss function with respect to all weights and biases to enable weight updates                                                                                                               |
| **Direction of Flow**            | Forward: from input layer → hidden layers → output layer                                                                   | Backward: from output layer → hidden layers → input layer                                                                                                                                                            |
| **Phase in Training Cycle**      | First phase of each training iteration (before weight update). Used in both training and inference                         | Second phase of training iteration (after forward pass, before weight update). Only used during training                                                                                                             |
| **Occurs During**                | Training (to compute loss) and inference (to generate predictions)                                                         | Exclusively during training (never in inference/prediction phase)                                                                                                                                                    |
| **Input Required**               | Input features (X) and current network parameters (weights/biases)                                                         | Output of forward pass (predictions, activations) and the loss value/target labels (y)                                                                                                                               |
| **Output Produced**              | Activations at each layer, final prediction ŷ, and loss value L = Loss(ŷ, y)                                               | Gradients ∂L/∂w for every weight w and ∂L/∂b for every bias b in the network                                                                                                                                         |
| **Loss Function Involvement**    | Directly computes the loss by comparing prediction to ground truth                                                         | Uses the derivative of the loss function w.r.t. the output to start the backward pass                                                                                                                                |
| **Weight Updates**               | No weight updates occur. Weights remain unchanged                                                                          | Provides gradients needed for weight updates (via optimizer like SGD, Adam, etc.)                                                                                                                                    |
| **Mathematical Foundation**      | Repeated application of linear transformation + activation: $ z^l = W^l a^{l-1} + b^l $, $ a^l = \sigma(z^l) $             | Chain rule of calculus applied layer-by-layer: $ \frac{\partial L}{\partial W^l} = \frac{\partial L}{\partial a^l} \odot \left( \frac{\partial a^l}{\partial z^l} \right) \odot (a^{l-1})^T $ (with local gradients) |
| **Chain Rule Application**       | Not required (simple feed-forward computation)                                                                             | Core mechanism: gradients are computed by multiplying local derivatives backward through the computation graph                                                                                                       |
| **Error Signal**                 | No explicit error signal; only final loss is computed                                                                      | Error signal $ (\delta^l = \frac{\partial L}{\partial z^l}) $ is propagated backward: $ \delta^l = (W^{l+1})^T \delta^{l+1} \odot \sigma'(z^l) $                                                                     |
| **Activation Storage**           | Must store activations a^l and pre-activations z^l for all layers (needed for backprop)                                    | Reuses stored activations and pre-activations from forward pass to compute gradients efficiently                                                                                                                     |
| **Computational Complexity**     | $ O\left( \sum*{l} N_l \times N*{l-1} \right) $ where $ N_l $ is number of neurons in layer l (linear in total parameters) | Roughly 2× the cost of forward pass (due to gradient computations for inputs and weights)                                                                                                                            |
| **Memory Requirements**          | Moderate: store activations and pre-activations for all layers and batch                                                   | High: same as forward + storage for gradients of all parameters; peak memory usually during backprop                                                                                                                 |
| **Numerical Stability Concerns** | Sensitive to activation function choice (e.g., sigmoid saturation)                                                         | Additional concerns: vanishing/exploding gradients (mitigated by ReLU, BatchNorm, Gradient clipping)                                                                                                                 |
| **Parallelization**              | Highly parallelizable across batch and layers (easy on GPUs)                                                               | Also highly parallelizable, but requires careful synchronization of gradients (e.g., in distributed training)                                                                                                        |
| **Use in Inference**             | Essential: the only computation needed to generate predictions                                                             | Not used: inference runs only forward propagation for speed and efficiency                                                                                                                                           |
| **Role in Optimization**         | Provides the loss value that drives optimization                                                                           | Provides the direction (gradients) for parameter updates via optimizers (SGD, RMSprop, Adam, etc.)                                                                                                                   |
| **Variants/Extensions**          | Can be modified for attention mechanisms, residual connections, etc                                                        | Backprop through time (BPTT) for RNNs, truncated backprop, gradient checkpointing for memory efficiency                                                                                                              |
| **Historical Context**           | Core idea present since early neural networks (rumelhart et al. formalized with backprop)                                  | Introduced in 1986 by Rumelhart, Hinton, Williams; enabled training of multi-layer networks                                                                                                                          |
| **Use Cases**                    | Fast and predictable latency; critical for real-time serving systems (e.g., recommendation engines)                        | Memory-intensive; often the bottleneck in large-scale training (use mixed-precision, gradient accumulation)                                                                                                          |


# Convolutional Neural Networks (CNNs)
