# Learning Types

| Aspect            | Supervised                                                                                        | Unsupervised                                                                                          | Semi-Supervised                                                                          | Reinforcement                                                                                                  | Self-Supervised                                                                                                                                | Online                                                                                                                                   | Transfer                                                                                                                   | Ensemble                                                                                       |
| ----------------- | ------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------- |
| Description       | Model learns from input-output pairs to map inputs to known outputs                               | Model finds hidden patterns or structures in data without any labels                                  | Combines a small amount of labeled data with a large amount of unlabeled data            | Agent learns optimal actions by trial-and-error interaction with an environment, guided by rewards/punishments | A subset of unsupervised learning where supervisory signals (pseudo-labels) are automatically generated from the data itself via pretext tasks | Model learns incrementally from a continuous stream of data, updating after each example or small batch (contrasted with batch learning) | Knowledge (usually a pretrained model) from a source task/domain is reused/adapted for a different but related target task | Combines predictions from multiple models to achieve better performance than individual models |
| Data Requirements | Fully labeled data (inputs + correct outputs)                                                     | Unlabeled data                                                                                        | Small labeled + large unlabeled                                                          | States, actions, rewards (no direct labels)                                                                    | Unlabeled data (pretext tasks create labels)                                                                                                   | Streaming data (sequential arrival)                                                                                                      | Pretrained model + target task data (often small)                                                                          | Depends on base models                                                                         |
| Supervision Level | Full supervision                                                                                  | No supervision                                                                                        | Partial supervision                                                                      | No direct supervision                                                                                          | No external supervision                                                                                                                        | Can be any (supervised/unsupervised/RL)                                                                                                  | Depends on source/target                                                                                                   | Depends on base models                                                                         |
| Labeled Data      | Yes                                                                                               | No                                                                                                    | Partial                                                                                  | No                                                                                                             | No (auto-labels)                                                                                                                               | Optional                                                                                                                                 | Pretrained                                                                                                                 | Depends                                                                                        |
| Feedback          | Direct labels                                                                                     | Patterns                                                                                              | Mixed                                                                                    | Rewards                                                                                                        | Internal                                                                                                                                       | Continuous                                                                                                                               | Reused knowledge                                                                                                           | Aggregated predictions                                                                         |
| Objective         | Minimize prediction error on labeled data                                                         | Discover intrinsic data structure                                                                     | Leverage unlabeled data to improve performance                                           | Maximize cumulative long-term reward                                                                           | Learn useful representations via pretext tasks                                                                                                 | Adapt to new data quickly without retraining                                                                                             | Accelerate and improve learning on target task                                                                             | Improve accuracy and robustness                                                                |
| Key Goal          | Predict outputs                                                                                   | Find hidden patterns                                                                                  | Combine both learning types                                                              | Learn best actions                                                                                             | Learn data representations                                                                                                                     | Learn incrementally                                                                                                                      | Transfer knowledge                                                                                                         | Combine predictions                                                                            |
| Common Algorithms | Linear/Logistic Regression, SVM, Decision Trees, Random Forests, Neural Networks                  | K-Means, Hierarchical Clustering, PCA, t-SNE, Autoencoders, GMM, DBSCAN                               | Self-training, Co-training, Label Propagation, Graph-based methods, S3VM                 | Q-Learning, SARSA, DQN, Policy Gradients, Actor-Critic (A2C/A3C), PPO, TD3                                     | Contrastive methods (SimCLR, MoCo), Masked modeling (BERT, MAE), Rotation prediction, Jigsaw puzzles                                           | Stochastic Gradient Descent (SGD), Online K-Means, Perceptron, Hoeffding Trees, Incremental PCA                                          | Fine-tuning (BERT, ResNet, GPT), Feature extraction, Domain adaptation techniques                                          | Bagging (Random Forest), Boosting (AdaBoost, XGBoost), Stacking                                |
| Advantages        | High accuracy with sufficient data; clear evaluation metrics; direct optimization                 | No labeling needed; useful for exploratory analysis; handles large unlabeled datasets                 | Reduces labeling cost; often better performance than pure supervised with limited labels | Handles sequential decision-making; learns from interaction; no need for labeled dataset                       | Scales to massive unlabeled data; produces strong general-purpose representations; powers foundation models                                    | Memory-efficient; handles non-stationary data; suitable for real-time; no need to store all data                                         | Reduces data and compute needs for target; leverages large pretraining datasets; often state-of-the-art performance        | Better performance, reduced overfitting, robustness                                            |
| Disadvantages     | Requires large amounts of labeled data; labeling is expensive/time-consuming; risk of overfitting | Results hard to evaluate/interpret; no guarantee of meaningful patterns; sensitive to hyperparameters | Relies on assumptions (e.g., cluster or manifold assumption); risk of propagating errors | Sample-inefficient (needs many interactions); reward design is critical; exploration-exploitation challenge    | Requires designing effective pretext tasks; downstream fine-tuning often needed                                                                | Risk of catastrophic forgetting; sensitive to order of data; harder to achieve convergence                                               | Risk of negative transfer (if tasks unrelated); requires relevant source task; domain shift issues                         | Higher computational cost, harder to interpret                                                 |
| Use Cases         | Classification (e.g., spam detection), Regression (e.g., price prediction)                        | Clustering (customer segmentation), Dimensionality reduction, Anomaly detection                       | Image/text classification with limited labels, Medical imaging, Web page classification  | Game playing (AlphaGo), Robotics, Autonomous driving, Resource management                                      | Pretraining for NLP (BERT/GPT) and Vision (ViT, DINO), Representation learning                                                                 | Recommendation systems, Fraud detection, IoT sensor data, Stock price prediction                                                         | Computer vision (ImageNet → medical images), NLP (pretrained LM → specific downstream), Cross-domain adaptation            | Classification, Regression, Anomaly detection, Various domains                                 |
