# Matrix

**Matrix**: structured grid of numbers arranged in rows and columns, used to represent and manipulate data.


## Notation and Nomenclature

| Symbol                                                                                   | Meaning                                                                           |
| ---------------------------------------------------------------------------------------- | --------------------------------------------------------------------------------- |
| $(a, b, c)$                                                                              | Scalars (usually lowercase italic)                                                |
| $(\alpha, \beta, \gamma, \theta, \lambda, \sigma)$                                       | Scalars (Greek letters; parameters, eigenvalues, singular values, learning rates) |
| $(\mathbf{A}, \mathbf{B}, \mathbf{X}, \mathbf{Y}, \mathbf{W}, \mathbf{H})$               | Matrices (uppercase bold / blackboard bold)                                       |
| $(\mathbb{R}, \mathbb{C})$                                                               | Real and complex numbers                                                          |
| $(\mathbb{R}^n, \mathbb{R}^{m \times n})$                                                | Space of (n)-vectors, $(m \times n)$ matrices                                     |
| $(\mathbf{a}^\top, \mathbf{A}^\top)$                                                     | Transpose                                                                         |
| $(\mathbf{a}^\mathrm{H}, \mathbf{A}^\mathrm{H})$                                         | Conjugate (Hermitian) transpose                                                   |
| $(\mathbf{A}^{-1})$                                                                      | Matrix inverse                                                                    |
| $(\mathbf{A}^+)$                                                                         | Moore–Penrose pseudoinverse                                                       |
| $(\det(A))$ or $(\lVert A \rVert)$                                                       | Determinant                                                                       |
| $(\operatorname{tr}(A))$ or $(\mathrm{trace}(A))$                                        | Trace                                                                             |
| $(\operatorname{rank}(A))$                                                               | Rank                                                                              |
| $(\operatorname{diag}(\mathbf{a}))$ or $(\operatorname{diag}(A))$                        | Diagonal matrix or vector of diagonal entries                                     |
| $(\operatorname{vec}(A))$                                                                | Vectorization (stack columns)                                                     |
| $(\lVert \mathbf{a} \rVert_2)$ or $(\lVert \mathbf{a} \rVert)$                           | Euclidean (L2) norm                                                               |
| $(\lVert \mathbf{a} \rVert_1,\ \lVert \mathbf{a} \rVert_p,\ \lVert \mathbf{a} \rVert_0)$ | L1, Lp, L0 (pseudo) norms                                                         |
| $(\lVert A \rVert_2,\ \lVert A \rVert_F,\ \lVert A \rVert_*)$                            | Operator, Frobenius, nuclear norms                                                |
| $(\langle \mathbf{a}, \mathbf{b} \rangle)$ or $(\mathbf{a} \cdot \mathbf{b})$            | Inner (dot) product                                                               |

## Operations with Tensors

### Simple Mathematical Operations

- **Addition**: Element-wise addition of two tensors of the same shape. For tensors $\mathcal{A}$ and $\mathcal{B}$, the sum $\mathcal{C} = \mathcal{A} + \mathcal{B}$ is defined as $c_{ijk} = a_{ijk} + b_{ijk}$ = $\begin{bmatrix} a_{1} \\ a_{2} \end{bmatrix} + \begin{bmatrix} b_{1} \\ b_{2} \end{bmatrix} = \begin{bmatrix} a_{1} + b_{1} \\ a_{2} + b_{2} \end{bmatrix}$
- **Subtraction**: Element-wise subtraction of two tensors of the same shape. For tensors $\mathcal{A}$ and $\mathcal{B}$, the difference $\mathcal{C} = \mathcal{A} - \mathcal{B}$ is defined as $c_{ijk} = a_{ijk} - b_{ijk}$ = $\begin{bmatrix} a_{1} \\ a_{2} \end{bmatrix} - \begin{bmatrix} b_{1} \\ b_{2} \end{bmatrix} = \begin{bmatrix} a_{1} - b_{1} \\ a_{2} - b_{2} \end{bmatrix}$
- **Scalar Multiplication**: Multiplying each element of a tensor by a scalar. For a tensor $\mathcal{A}$ and scalar $\alpha$, the result $\mathcal{B} = \alpha \mathcal{A}$ is defined as $b_{ijk} = \alpha \cdot a_{ijk}$ = $\alpha \begin{bmatrix} a_{1} \\ a_{2} \end{bmatrix} = \begin{bmatrix} \alpha \cdot a_{1} \\ \alpha \cdot a_{2} \end{bmatrix}$
- **Multiplication**: Element-wise multiplication of two tensors of the same shape. For tensors $\mathcal{A}$ and $\mathcal{B}$, the product $\mathcal{C} = \mathcal{A} \circ \mathcal{B}$ is defined as $c_{ijk} = a_{ijk} \cdot b_{ijk}$ = $\begin{bmatrix} a_{1} \\ a_{2} \end{bmatrix} \circ \begin{bmatrix} b_{1} \\ b_{2} \end{bmatrix} = \begin{bmatrix} a_{1} \cdot b_{1} \\ a_{2} \cdot b_{2} \end{bmatrix}$
- **Division**: Element-wise division of two tensors of the same shape. For tensors $\mathcal{A}$ and $\mathcal{B}$, the quotient $\mathcal{C} = \mathcal{A} \div \mathcal{B}$ is defined as $c_{ijk} = a_{ijk} / b_{ijk}$ = $\begin{bmatrix} a_{1} \\ a_{2} \end{bmatrix} \div \begin{bmatrix} b_{1} \\ b_{2} \end{bmatrix} = \begin{bmatrix} a_{1} / b_{1} \\ a_{2} / b_{2} \end{bmatrix}$

### Common Tensor Operations

- **Tensor Contraction**: Reducing the order of a tensor by summing over one or more pairs of indices:
  - **vector**: dot product: $\mathbf{a} \cdot \mathbf{b} = \sum_{i} a_{i} b_{i}$ = $a_{1}b_{1} + a_{2}b_{2} + ... + a_{n}b_{n}$
  - **matrix**: matrix multiplication: $\mathbf{C} = \mathbf{A} \cdot \mathbf{B}$ where $c_{ij} = \sum_{k} a_{ik} b_{kj}$ = $\begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix} \cdot \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix} = \begin{bmatrix} a_{11}b_{11} + a_{12}b_{21} & a_{11}b_{12} + a_{12}b_{22} \\ a_{21}b_{11} + a_{22}b_{21} & a_{21}b_{12} + a_{22}b_{22} \end{bmatrix}$
  - **higher-order tensors**: generalization of matrix multiplication, summing over shared indices: $U^{ac} = \sum_{b} T^{abc} V_{b} = \sum_{b} \begin{pmatrix} T^{11b} & T^{12b} \\ T^{21b} & T^{22b} \end{pmatrix} V_b = \sum_{b} \begin{pmatrix} 1 & 3 \\ 5 & 7 \end{pmatrix} V_1 + \sum_{b} \begin{pmatrix} 2 & 4 \\ 6 & 8 \end{pmatrix} V_2 = \begin{pmatrix} 1 \cdot 1 + 2 \cdot 2 & 3 \cdot 1 + 4 \cdot 2 \\ 5 \cdot 1 + 6 \cdot 2 & 7 \cdot 1 + 8 \cdot 2 \end{pmatrix} = \begin{pmatrix} 5 & 11 \\ 17 & 23 \end{pmatrix}$
- **Reshaping**: Changing the shape of a tensor without altering its data. For example, a tensor of shape (2 - rows, 3 - columns, 4 - depth) can be reshaped to (6, 4) or (2, 12), provided the total number of elements remains the same
- **Transposition**: Swapping (transpose) the axes of a tensor. For a 2D matrix (which is a 2D tensor), transposition swaps rows and columns: $\mathbf{A}^\top$ where $a_{ij}^\top = a_{ji}$ = $\begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix}^\top = \begin{bmatrix} a_{11} & a_{21} \\ a_{12} & a_{22} \end{bmatrix}$
- **Inversion**: Finding a tensor that, when multiplied with the original tensor, yields the identity tensor. For a square matrix (2D tensor), the inverse $\mathbf{A}^{-1}$ satisfies $\mathbf{A} \cdot \mathbf{A}^{-1} = \mathbf{I}$ = $\begin{bmatrix} a & b \\ c & d \end{bmatrix}^{-1} = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$ (if $ad - bc \neq 0$)
- **Identity Tensor**: A tensor that acts as a neutral element in tensor multiplication. For a 2D matrix (2D tensor), the identity matrix $\mathbf{I}$ has ones on the diagonal and zeros elsewhere: $\mathbf{I} = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$
- **Determinant**: A scalar value that can be computed from the elements of a square matrix (2D tensor), providing important properties about the matrix, such as whether it is invertible. The determinant of a matrix $\mathbf{A}$ is denoted as $\det(\mathbf{A})$ = $ad - bc$ for a 2x2 matrix $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$
- **Trace**: The sum of the diagonal elements of a square matrix (2D tensor). The trace of a matrix $\mathbf{A}$ is denoted as $\operatorname{tr}(\mathbf{A})$ = $a_{11} + a_{22} + ... + a_{nn}$ for an $n \times n$ matrix
- **Rank**: The dimension of the vector space generated by the columns (or rows) of the matrix (2D tensor). The rank of a matrix $\mathbf{A}$ is denoted as $\operatorname{rank}(\mathbf{A})$
- **Slicing**: Extracting a sub-tensor from a larger tensor by specifying ranges for each dimension. For example, slicing a 3D tensor $\mathcal{T}$ to get a 2D slice: $\mathcal{S} = \mathcal{T}[:, :, k]$
- **Broadcasting**: A method to perform operations on tensors of different shapes by automatically expanding the smaller tensor along the dimensions of the larger tensor. For example, adding a vector to each row of a matrix
- **Element-wise Operations**: Applying functions to each element of a tensor independently, such as applying the ReLU function: $f(x) = \max(0, x)$ to each element of tensor $\mathcal{A}$ to produce tensor $\mathcal{B}$
- **Norms**: Measures of the size or length of a tensor. Common norms include the Frobenius norm for matrices and the L2 norm for vectors. For a tensor $\mathcal{A}$, the Frobenius norm is defined as $\|\mathcal{A}\|_F = \sqrt{\sum_{i,j,k} a_{ijk}^2}$ = $\sqrt{a_{1}^2 + a_{2}^2 + ... + a_{n}^2}$
- **Decomposition**: Breaking down a tensor into simpler components, such as CANDECOMP/PARAFAC (CP) decomposition or Tucker decomposition, which are used in data analysis and compression
- **Convolution**: A specialized operation often used in signal processing and neural networks, where a kernel (smaller tensor) is slid over a larger tensor to produce a feature map. For a 2D tensor (image) and a 2D kernel, the convolution operation is defined as:
  $(\mathcal{I} * \mathcal{K})(i,j) = \sum_{m}\sum_{n} \mathcal{I}(i+m, j+n) \cdot \mathcal{K}(m,n)$ where $\mathcal{I}$ is the input tensor and $\mathcal{K}$ is the kernel tensor
