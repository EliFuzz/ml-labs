# Vector

**Vector**: ordered collection of numbers (elements) that represent a point or direction in a multi-dimensional space. In machine learning, vectors are used to represent data points, features, weights, and more.

## Notation and Nomenclature

| Notation                                 | Meaning                                              |
| ---------------------------------------- | ---------------------------------------------------- |
| $\mathbf{x}$ or $\vec{x}$                | Vector name is bold lowercase or with arrow overhead |
| $x_i$                                    | i-th component of vector $\mathbf{x}$                |
| $x^{(i)}$                                | i-th training example vector                         |
| $x^{([i])}$ or $x[i]$                    | i-th element using 0-based indexing                  |
| $\mathbf{x}_{1:n}$                       | Subvector from 1 to n                                |
| $\mathbf{1}$                             | All-ones vector                                      |
| $\mathbf{0}$                             | Zero vector                                          |
| $\mathbf{w}$, $\boldsymbol{\theta}$      | Weight / parameter vector                            |
| $\mathbf{b}$                             | Bias vector (sometimes scalar)                       |
| $\mathbf{h}$, $\mathbf{a}$, $\mathbf{z}$ | Hidden, activation, pre-activation                   |
| $\boldsymbol{\mu}$                       | Mean vector                                          |
| $\mathbf{q}$, $\mathbf{k}$, $\mathbf{v}$ | Query, key, value vectors                            |
| $\mathbf{x}^T$                           | Transpose (column → row)                             |
| $\mathbf{x}^H$                           | Hermitian (conjugate) transpose                      |
| $\|\mathbf{x}\|_0$                       | L0 pseudo-norm (sparsity count)                      |
| $\|\mathbf{x}\|_1$                       | L1 (Manhattan) norm                                  |
| $\|\mathbf{x}\|$                         | Euclidean (L2) norm by default                       |
| $\|\mathbf{x}\|_2$                       | Explicit L2 norm                                     |

## Dot Product (Inner Product)

### Definition

- **Dot Product**: A scalar value that represents the product of the magnitudes of two vectors and the cosine of the angle between them (measures how much two vectors point in the same direction). In ML, uses to measure similarity, compute weighted sums (neurons), project vectors, and calculate attention weights.

Example:

- **Formula**: $\begin{bmatrix} u_1 & u_2 \end{bmatrix} \cdot \begin{bmatrix} w_1 & w_2 \end{bmatrix} = u_1 \times w_1 + u_2 \times w_2 = u_1 w_1 + u_2 w_2 = \mathbf{u} \cdot \mathbf{w}$
- **Dot Product:** $\begin{bmatrix} 3 \\ 4 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ -2 \end{bmatrix} = 3 \times 1 + 4 \times (-2) = 3 + (-8) = -5$

### Notation

- **Vector**: $\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^{n} (u_i \times v_i)$
- **Geometric**: $\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \times \|\mathbf{v}\| \times \cos(\theta)$
- **Matrix**: $\mathbf{u} \cdot \mathbf{v} = \mathbf{u}^T \mathbf{v}$

### Properties

- **Commutative**: the order of the vectors does not change the result: $\mathbf{u} \cdot \mathbf{v} = \mathbf{v} \cdot \mathbf{u}$
- **Distributive**: the dot product distributes over vector addition: $\mathbf{u} \cdot (\mathbf{v} + \mathbf{w}) = \mathbf{u} \cdot \mathbf{v} + \mathbf{u} \cdot \mathbf{w}$
- **Scalar Multiplication**: multiplying a vector by a scalar scales the dot product: $(c \mathbf{u}) \cdot \mathbf{v} = c (\mathbf{u} \cdot \mathbf{v})$
- **Orthogonality**: if the dot product is zero, the vectors are orthogonal (perpendicular): $\mathbf{u} \cdot \mathbf{v} = 0 \implies \theta = 90^\circ$
- **Same Direction**: if the vectors point in the same direction, the dot product equals the product of their magnitudes: $\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \times \|\mathbf{v}\|$
- **Opposite Direction**: if the vectors point in opposite directions, the dot product equals the negative product of their magnitudes: $\mathbf{u} \cdot \mathbf{v} = -\|\mathbf{u}\| \times \|\mathbf{v}\|$

### Interpretations

| Context                   | What the dot product actually computes                                                                                                                                                  | Use Cases                                                                           |
| ------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------- |
| **Similarity measure**    | How aligned 2 vectors are (cosine similarity = dot product after normalization)                                                                                                         | Recommendation systems, word embeddings, sentence similarity                        |
| **Linear transformation** | $\mathbf{w} \cdot \mathbf{x} = $ linear combination of features $\mathbf{x}$ with weights $\mathbf{w}$                                                                                  | Single neuron output (before activation) in neural nets                             |
| **Projection**            | $\mathrm{proj}_{\mathbf{v}}(\mathbf{u}) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{v}\|^2} \mathbf{v}$                                                                              | Feature importance, attention scores                                                |
| **Attention mechanism**   | $\mathbf{Query} \cdot \mathbf{Key} \implies$ attention weight                                                                                                                           | Transformers (scaled dot-product attention)                                         |
| **Loss functions**        | Many losses contain dot products (e.g., contrastive loss, triplet loss)                                                                                                                 | Self-supervised learning (SimCLR, MoCo)                                             |
| **Kernel trick**          | $K(\mathbf{x},\mathbf{y}) = \phi(\mathbf{x}) \cdot \phi(\mathbf{y})$ in high-dimensional space                                                                                          | SVMs (Support Vector Machines) with RBF (Radial Basis Function), polynomial kernels |
| **Gradient computation**  | $\frac{\partial}{\partial \mathbf{w}} \left( \frac{1}{2} \|\mathbf{y} - \mathbf{w} \cdot \mathbf{x}\|^2 \right)$ contains $(\mathbf{w} \cdot \mathbf{x} - \mathbf{y}) \cdot \mathbf{x}$ | Linear regression, logistic regression update rule                                  |

### Common Cases

| Dot product value       | Angle θ | Interpretation                                                                                                                |
| ----------------------- | ------- | ----------------------------------------------------------------------------------------------------------------------------- |
| Positive large          | `< 90°` | Vectors are similar → high attention weight, similar embeddings                                                               |
| `≈ 0`                   | `≈ 90°` | Unrelated / orthogonal features. Whitening / decorrelation of features (removes multicollinearity → more stable coefficients) |
| Negative large          | `> 90°` | Dissimilar → pushed apart in contrastive learning                                                                             |
| `= \|\|u\|\|×\|\|v\|\|` | `0°`    | Identical direction (after normalization → similarity = 1)                                                                    |

### Examples

| Case                                | Dot Product Behavior                  | Appears in                                                  |
| ----------------------------------- | ------------------------------------- | ----------------------------------------------------------- |
| Weight × Input (w · x)              | Output of a neuron before activation  | Every linear layer, fully-connected, convolution (unrolled) |
| Embedding × Embedding               | Semantic similarity score             | Word2vec skip-gram, BERT pooling                            |
| Query × Key                         | Raw attention score                   | All Transformer models                                      |
| Gradient of MSE(Mean Squared Error) | Contains (prediction - label) · input | Backpropagation in linear models                            |

| Norm               | Formula                                                                                             | Definition                                                                                                  | Use Cases                                                                                                                                                                                       |
| ------------------ | --------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| ℓᵖ                 | $\|\mathbf{x}\|_p = \left( \sum_{i=1}^n \|x_i\|^p \right)^{\frac{1}{p}}$ where $p \geq 1$           | measures the "length" of vectors in finite-dimensional spaces                                               | Various applications depending on p value                                                                                                                                                       |
| No Normalization   | $u \cdot v$                                                                                         | Raw vector                                                                                                  | Neural network layers, attention before softmax                                                                                                                                                 |
| L1 (Manhattan)     | $\|\mathbf{v}\|_1 = \sum_{i=1}^{n} \|v_i\|$                                                         | Euclidean distance from the origin to the point identiﬁed by x. Sum of absolute values of vector components | Sparse representations, Lasso regression                                                                                                                                                        |
| L2 (Euclidean)     | $\|\mathbf{v}\|_2 = \sqrt{\sum_{i=1}^{n} v_i^2}$                                                    | Square root of sum of squares of components                                                                 | Attention weights in Transformers (query/key dot-products → normalize to unit length to stabilize gradients); Embedding normalization (use unit embeddings for faster & more stable ANN search) |
| L3                 | $\|\mathbf{v}\|_3 = \left( \sum_{i=1}^{n} \|v_i\|^3 \right)^{1/3} = \sqrt[3]{\sum_{i=1}^{n} v_i^3}$ | Generalization of Lp norms                                                                                  | Less common, used in specific applications                                                                                                                                                      |
| Infinity           | $\|\mathbf{v}\|_\infty = \max(\|v_1\|, \|v_2\|, ..., \|v_n\|)$                                      | Maximum absolute value among components                                                                     | Chebyshev distance, robust optimization                                                                                                                                                         |
| Scaled dot-product | $\frac{u \cdot v}{\sqrt{d_k}}$                                                                      | Dot product scaled by dimension of vectors                                                                  | Transformer attention mechanism (prevents vanishing gradients with high dim)                                                                                                                    |

### Vector Types

- **Unit Vectors**: vectors with a magnitude of 1. Used to indicate direction without affecting magnitude ("Cares about direction, not magnitude"). Example: $\mathbf{u} = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}$ has a magnitude of 1.
  - **Unit length (magnitude)**: calculated as the square root of the sum of its squared components: $\|\mathbf{u}\| = \sqrt{u_1^2 + u_2^2 + ... + u_n^2} = l$
- **Orthonormal Vectors**: vectors that are both orthogonal (perpendicular) to each other and have a magnitude of 1. Example: $\mathbf{u} \cdot \mathbf{v} = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \cdot \begin{bmatrix} 0 \\ 1 \end{bmatrix} = 0$ and both have magnitude 1.
- **Orthogonal Vectors**: vectors that are perpendicular to each other, meaning their dot product is zero. Example: $\mathbf{u} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$ and $\mathbf{v} = \begin{bmatrix} -2 \\ 1 \end{bmatrix}$ have a dot product of zero: $\mathbf{u} \cdot \mathbf{v} = 1 \times -2 + 2 \times 1 = 0$
- **Eigenvectors**: non-zero vectors that only scale (not change direction) when a linear transformation (like a matrix) is applied to them. Example: For matrix $\mathbf{A} = \begin{bmatrix} 2 & 0 \\ 0 & 3 \end{bmatrix}$, eigenvectors are $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ and $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$ with eigenvalues 2 and 3 respectively.
  - **Eigenvalue**: scalar factor by which that eigenvector is scaled (stretched, shrunk, or flipped) during the transformation: $\mathbf{A} \mathbf{v} = \lambda \mathbf{v}$

### Vector Projection

- **Projection of u onto v**: $\mathrm{proj}_{\mathbf{v}}(\mathbf{u}) = \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{v}\|^2} \mathbf{v}$
- **Example**: Projecting vector $\mathbf{u} = \begin{bmatrix} 3 \\ 2 \end{bmatrix}$ onto vector $\mathbf{v} = \begin{bmatrix} 4 \\ 0 \end{bmatrix}$:
  - Calculate dot product: $\mathbf{u} \cdot \mathbf{v} = 3 \times 4 + 2 \times 0 = 12$
  - Calculate squared magnitude of v: $\|\mathbf{v}\|^2 = 4^2 + 0^2 = 16$
  - Compute scalar projection: $\frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{v}\|^2} = \frac{12}{16} = 0.75$
  - Final projection vector: $0.75 \times \begin{bmatrix} 4 \\ 0 \end{bmatrix} = \begin{bmatrix} 3 \\ 0 \end{bmatrix}$


```python
from matplotlib.pyplot import (
    figure,
    axhline,
    axvline,
    grid,
    arrow,
    legend,
    title,
    show,
    xlim,
    ylim,
)
from numpy import array


def plot_vectors(labels, vectors, colors, xlims=(-5, 5), ylims=(-5, 5)):
    figure(figsize=(6, 6))
    axhline(0, color="#8E8E93", lw=1)
    axvline(0, color="#8E8E93", lw=1)
    grid(True, color="#8E8E93", alpha=0.3)

    for vec, color, label in zip(vectors, colors, labels):
        arrow(
            0,
            0,
            vec[0],
            vec[1],
            head_width=0.2,
            head_length=0.3,
            fc=color,
            ec=color,
            linewidth=2,
            label=label,
        )

    xlim(xlims)
    ylim(ylims)
    legend()
    title(label)
    show()
```

```python
a = array([2, 1])
b = array([1, 3])

plot_vectors(
    ["a", "b", "a+b", "2*a", "-a"],
    [a, b, a + b, 2 * a, -a],
    ["#0088FF", "#FF383C", "#34C759", "#CB30E0", "#FF8D28"],
)
```

```python
from numpy import array, dot, linalg, arccos, degrees, clip

a = array([1, 2])
b = array([2, 1])

cosine_sim = dot(a, b) / (linalg.norm(a) * linalg.norm(b))
angle_rad = arccos(clip(cosine_sim, -1, 1))
angle_deg = degrees(angle_rad)

plot_vectors(
    [f"a (Cosine similarity: {cosine_sim:.2f}. Angle: {angle_deg:.1f}°)", "b"],
    [a, b],
    ["blue", "red"],
    xlims=(0, 3),
    ylims=(0, 3),
)
```

```python
from numpy import array, dot

a = array([3, 2])
b = array([4, 0])

plot_vectors(
    ["a (to project)", "b (axis)", "projection of a onto b"],
    [a, b, (dot(a, b) / dot(b, b)) * b],
    ["blue", "red", "green"],
    xlims=(0, 5),
    ylims=(-1, 4),
)
```
