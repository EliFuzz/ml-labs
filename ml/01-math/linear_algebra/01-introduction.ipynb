# Overview


## Core Concepts

- **Scalar**: A single value, such as a number: $\alpha$, 42
- **Vector**: An ordered list of values, represented as a column: $\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix}$
- **Matrix**: A two-dimensional array of values, represented as a grid: $\mathbf{M} = \begin{bmatrix} m_{11} & m_{12} \\ m_{21} & m_{22} \end{bmatrix}$
- **Tensor**: A multi-dimensional array of values, generalizing scalars, vectors, and matrices to higher dimensions. 3D tensor: $\mathcal{T} = \begin{bmatrix} \begin{bmatrix} t_{111} & t_{112} \\ t_{121} & t_{122} \end{bmatrix} \\ \begin{bmatrix} t_{211} & t_{212} \\ t_{221} & t_{222} \end{bmatrix} \end{bmatrix}$

## Product Types

| Aspect                                            | Inner Product              | Matrix Product                  | Hadamard Product            |
| ------------------------------------------------- | -------------------------- | ------------------------------- | --------------------------- |
| **Operation Type**                                | Dot product                | Row-column multiplication       | Element-wise multiplication |
| **Inputs**                                        | Two vectors of same length | Two matrices (inner dims match) | Two arrays of same shape    |
| **Output**                                        | Scalar                     | Matrix                          | Array of same shape         |
| **Commutative (order doesn't affect the result)** | Yes                        | No                              | Yes                         |
| **Requires same shape**                           | Yes                        | No                              | Yes                         |

- **Inner Product**: $\mathbf{u}\cdot\mathbf{v} = \sum_{i=1}^{n} u_i v_i = u_1v_1 + u_2v_2 + \dots + u_n v_n = \begin{bmatrix} u_1 \\ u_2 \\ u_3 \end{bmatrix} \cdot \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} = u_1v_1 + u_2v_2 + u_3v_3 = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} \cdot \begin{bmatrix} 4 \\ 5 \\ 6 \end{bmatrix} = 1\cdot4 + 2\cdot5 + 3\cdot6 = 32$
- **Matrix Product**: $(AB)_{ij} = \sum_{k=1}^{n} A_{ik} B_{kj} = A_{i1}B_{1j} + A_{i2}B_{2j} + \dots + A_{in}B_{nj} =  \begin{bmatrix}1&2\\3&4\end{bmatrix} \begin{bmatrix}5&6\\7&8\end{bmatrix} = \begin{bmatrix} 1\cdot5 + 2\cdot7 & 1\cdot6 + 2\cdot8 \\ 3\cdot5 + 4\cdot7 & 3\cdot6 + 4\cdot8 \end{bmatrix} = \begin{bmatrix} 5+14 & 6+16\\ 15+28 & 18+32 \end{bmatrix} = \begin{bmatrix} 19 & 22\\ 43 & 50 \end{bmatrix}$
- **Hadamard Product**: $(A \circ B)_{ij} = A_{ij} B_{ij} = \begin{bmatrix}1&2\\3&4\end{bmatrix} \circ \begin{bmatrix}10&20\\30&40\end{bmatrix} = \begin{bmatrix} 1\cdot10 & 2\cdot20 \\ 3\cdot30 & 4\cdot40 \end{bmatrix} = \begin{bmatrix} 10 & 40\\ 90 & 160 \end{bmatrix}$
