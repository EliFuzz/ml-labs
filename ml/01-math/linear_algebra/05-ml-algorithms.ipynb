# Overview


## ML Algorithm Decision-Making Checklist

- **Business & Strategic Alignment**
  - **Business Outcomes & KPIs**:
    - What exact outcome must the model achieve?
    - Is it automation, optimization, prediction, or decision support?
    - What measurable KPIs define success (accuracy, speed, cost savings, risk reduction)?
  - **Business Value vs. Complexity**:
    - Does the complexity improve business value significantly?
    - Is a simpler approach sufficient?
    - Is explainability required for business users/executives?
  - **Regulatory & Domain Constraints**:
    - Industry-specific constraints (finance, healthcare, energy, defense)
    - Requirements for transparency (e.g., SOX, GDPR, PSD2)
  - **Time-to-Value**:
    - How quickly does the business need the first result?
    - Does your timeline favor simple models or advanced deep learning?
  - **Cost Constraints**:
    - Cost to build, train, deploy, operate, maintain
    - Required compute resources cost (GPU/TPU vs. CPU)
- **Data Considerations**
  - **Data Type**:
    - Structured (tabular), unstructured (text, images, audio), semi-structured
    - Mixed modality (tabular + text, etc.)
  - **Data Volume**:
    - Small dataset → simpler models (LR, SVM, tree-based)
    - Large dataset → deep learning, distributed training
  - **Data Quality**:
    - Noise levels
    - Missing values
    - Outliers
    - Label quality
  - **Data Distribution**:
    - Imbalanced classes
    - Skewed distributions
    - Seasonality
    - Non-stationarity (concept drift)
  - **Feature Space**:
    - Number of features (dimensionality)
    - Feature correlation
    - Sparse vs. dense features
    - Real-time availability of features (feature freshness)
  - **Labeling**:
    - Availability of labeled data
    - Label consistency
    - Cost and difficulty of producing labels
- **Problem Type Considerations**
  - **Supervised Learning**:
    - Regression
    - Classification
    - Ranking
    - Time series forecasting
  - **Unsupervised Learning**:
    - Clustering
    - Dimensionality reduction
    - Anomaly detection
  - **Semi-Supervised Learning**:
    - Few labels + many unlabeled samples
  - **Reinforcement Learning**:
    - Sequential decisions + reward optimization
  - **Generative Models**:
    - Content creation
    - Simulation
    - Synthetic data generation
- **Model Performance Requirements**
  - **Accuracy vs. Interpretability**:
    - What's more important: maximum accuracy or human understanding?
  - **Precision/Recall Tradeoff**:
    - Cost of false positives vs. false negatives
    - Domain-specific risk impacts (fraud, medical diagnosis)
  - **Speed Constraints**:
    - Training speed
    - Prediction latency (real-time vs. batch)
  - **Model Robustness**:
    - Sensitivity to noise
    - Sensitivity to adversarial attacks
    - Stability over time
  - **Scalability**:
    - Can the model scale to millions/billions of data points?
    - Distributed training capability (Spark, Ray, Horovod)
- **Operational & MLOps Considerations**
  - **Deployment Environment**:
    - On-prem
    - Hybrid
    - Cloud
    - Edge/IoT
    - Mobile
  - **Model Lifecycle Management**:
    - Versioning
    - Retraining automation
    - Model registry compatibility
  - **Resource Availability**:
    - GPU/TPU/CPU availability
    - Memory footprint
    - Storage constraints
  - **Monitoring & Observability**:
    - Feature drift monitoring
    - Model drift tracking
    - Data quality monitoring in production
  - **Integration Constraints**:
    - APIs
    - Microservices
    - Data pipelines
    - Real-time streaming
- **Security & Compliance Considerations**
  - **Explainability Requirements**:
    - Do regulators require transparent decisions?
    - Is black-box modeling allowed?
  - **Bias & Fairness**:
    - Protected attributes
    - Equal opportunity requirements
    - Bias testing needed?
  - **Data Permissioning**:
    - Can training data contain sensitive content?
    - Consent requirements (GDPR, CCPA)
  - **Privacy-Preserving Techniques**:
    - Federated learning
    - Differential privacy
    - Homomorphic encryption
- **Cost & ROI Considerations**
  - **Compute Cost**:
    - Training cost
    - Inference cost
    - Scaling costs
  - **Engineering Cost**:
    - Development time
    - Infrastructure setup
    - Ongoing maintenance
  - **ROI Horizon**:
    - Short-term vs. long-term benefits
    - Opportunity cost of complexity
- **Risk Management Considerations**
  - **Operational Risk**:
    - Probability of model failure
    - Impact of incorrect predictions
  - **Business Risk**:
    - Potential for adverse customer impact
    - Revenue or reputation risk
  - **Ethical & Social Risk**:
    - Algorithmic bias
    - Discrimination potential
  - **Model Drift Risk**:
    - Does the environment change often (dynamic vs. stable)?

## ML Algorithm Decision Lookup

- **Supervised Learning**: Learning from labeled data with labels
  - **Regression**: Predicting continuous values
    - **Linear & Parametric**: Linear or simple parametric forms (Linear Regression, Polynomial Regression, Ridge, Lasso, ElasticNet, Bayesian Regression, Quantile Regression, GLM)
    - **Tree-Based & Ensemble**: Tree and boosting models (Decision Tree Regressor, Random Forest Regressor, Extra Trees, XGBoost, LightGBM, CatBoost, NGBoost, HistGradientBoosting)
    - **Nearest Neighbors**: Instance-based regression (KNN Regressor)
    - **Kernel Methods**: Kernel-based non-linear regression (SVR, Gaussian Process Regression)
    - **Neural & Deep**: Deep learning regressors (MLP Regressor, TabNet, NODE, FT-Transformer, SAINT, TabTransformer)
  - **Classification**: Predicting discrete classes
    - **Linear & Probabilistic**: Linear or probabilistic models (Logistic Regression, Naive Bayes, LDA, QDA)
    - **Kernel Methods**: Margin-based classifiers (SVM / SVC)
    - **Nearest Neighbors**: Instance-based classification (KNN Classifier)
    - **Tree-Based & Ensemble**: Tree and gradient boosting classifiers (Decision Tree Classifier, Random Forest, Extra Trees, XGBoost, LightGBM, CatBoost)
    - **Neural & Deep**: Deep learning classifiers (MLP Classifier, Tabular transformers, CNNs/Transformers with classification heads)
- **Unsupervised Learning**: Finding patterns without labels
  - **Clustering**: Grouping similar points
    - **Centroid-Based**: Prototype-centered clusters (K-Means, K-Means++, Mini-Batch K-Means, K-Medoids)
    - **Hierarchical**: Tree-like nested clusters (Agglomerative Clustering (Ward, Complete, Average), BIRCH)
    - **Density-Based**: Density-connected regions (DBSCAN, HDBSCAN, OPTICS, MeanShift)
    - **Probabilistic**: Mixture models (Gaussian Mixture Models (GMM))
    - **Spectral & Graph**: Eigenvalue-based clustering (Spectral Clustering, Affinity Propagation)
    - **Deep Clustering**: Neural clustering methods (DEC, IDEC, VaDE, SCAN)
  - **Dimensionality Reduction**: Compressing feature space
    - **Linear Methods**: Linear projections (PCA, Kernel PCA, Incremental PCA, Truncated SVD, Factor Analysis, NMF)
    - **Non-Linear Manifold**: Non-linear embeddings (t-SNE, UMAP, Isomap, LLE, MDS, PHATE)
    - **Autoencoders**: Neural compression (Vanilla AE, Denoising AE, Sparse AE, Contractive AE, VAE, MAE)
  - **Anomaly Detection**: Identifying outliers (Isolation Forest, LOF, One-Class SVM, Elliptic Envelope, HBOS, COPOD, Autoencoder reconstruction error)
  - **Association Rule Mining**: Finding item co-occurrences (Apriori, FP-Growth, Eclat)
- **Semi-Supervised Learning**: Using few labels + lots of unlabeled data
  - **Wrapper Methods**: Iterative pseudo-labeling (Self-Training, Co-Training, Tri-Training)
  - **Graph-Based**: Label propagation on graphs (Label Propagation, Label Spreading, GNN-based SSL)
  - **Generative Models**: Joint distribution modeling (Semi-supervised VAE, EM with partial labels)
  - **Low-Density Separation**: Large margin in unlabeled regions (TSVM, S3VM)
  - **Consistency Regularization**: Invariance to augmentations (Π-Model, Mean Teacher, UDA, FixMatch, ReMixMatch, FlexMatch)
- **Self-Supervised Learning**: Learning representations via pretext tasks
  - **Contrastive Learning**: Positive/negative pair discrimination (SimCLR, MoCo v3, BYOL, SimSiam, SwAV, Barlow Twins, VICReg, DINO, iBOT)
  - **Masked Modeling**: Predict masked input (BERT, RoBERTa, MAE, BEiT, SimMIM, data2vec)
  - **Predictive Coding**: Predict future or context (GPT-style, CPC, SimCSE)
- **Reinforcement Learning**: Learning from rewards and interaction
  - **Model-Free Value-Based**: Learning Q/values (Q-Learning, SARSA, DQN, Double/Dueling/Rainbow DQN)
  - **Model-Free Policy-Based**: Direct policy gradients (REINFORCE, PPO, TRPO)
  - **Model-Free Actor-Critic**: Hybrid methods (A2C/A3C, PPO, SAC, TD3, DDPG)
  - **Model-Based**: Planning with learned dynamics (MBPO, DreamerV3, MuZero, World Models, PlaNet)
  - **Offline RL**: Learning from static datasets (Behavior Cloning, CQL, BCQ, TD3+BC, Decision Transformer, IQL)
- **Deep Learning Architectures**: Core neural network designs
  - **Feedforward**: Fully-connected networks (MLP, Deep Neural Networks)
  - **Convolutional**: Spatial hierarchies (ResNet, EfficientNet, ConvNeXt, MobileNet, DenseNet)
  - **Recurrent**: Sequential processing (RNN, LSTM, GRU)
  - **Transformer & Attention**: Self-attention models (BERT, GPT series, Llama, Mistral, Gemma, T5, RoBERTa)
  - **Vision Transformer**: Transformers for images (ViT, Swin Transformer, DeiT, BEiT)
  - **Graph Neural Networks**: Graph processing (GCN, GAT, GraphSAGE, GIN)
  - **Generative**: Data synthesis models (GANs, Diffusion Models, VAEs, Normalizing Flows)
- **Deep Learning Training Techniques**: Methods to train deep models effectively
  - **Optimizers**: Gradient descent variants (SGD, Adam, AdamW, RMSProp, LAMB)
  - **Regularization**: Preventing overfitting (Dropout, Weight Decay, BatchNorm, LayerNorm)
  - **Scaling Techniques**: Large-model training (Mixed Precision, Gradient Checkpointing, DDP, FSDP)
- **Natural Language Processing**: Understanding and generating text
  - **Classical Approaches**: Pre-transformer methods (Bag-of-Words, TF-IDF, Word2Vec, GloVe, FastText)
  - **Transformer-Based**: Large pretrained models (BERT, RoBERTa, XLNet, ELECTRA, DeBERTa, T5, GPT series, Llama, Mistral, Gemma, Phi)
  - **Advanced Techniques**: Reasoning enhancements (RAG, Chain-of-Thought, Tool Use, Agents)
- **Computer Vision**: Processing and understanding images/video
  - **CNN Evolution**: Convolutional backbones (AlexNet, VGG, ResNet, EfficientNet, ConvNeXt)
  - **Vision Transformers**: Image transformers (ViT, Swin Transformer, DeiT, BEiT, MAE)
  - **Detection & Segmentation**: Object/localization models (YOLO series, Faster R-CNN, Mask R-CNN, U-Net, SAM)
  - **Generative Vision**: Image synthesis (Stable Diffusion, DALL·E, Imagen, Sora-style)
- **Time Series Forecasting**: Predicting sequential data
  - **Classical Statistical**: Traditional models (ARIMA, SARIMA, Exponential Smoothing, Prophet)
  - **Tree-Based ML**: Boosted trees on lags (XGBoost, LightGBM, CatBoost)
  - **Deep Sequence Models**: Neural approaches (LSTM, GRU, TCN, Transformer-based (Informer, Autoformer, TFT), Mamba, N-BEATS, DeepAR, TiDE)
- **Ensemble & Automated ML**: Combining models & automation
  - **Bagging**: Bootstrap aggregating (Random Forest, Extra Trees, Bagged models)
  - **Boosting**: Sequential boosting (AdaBoost, Gradient Boosting, XGBoost, LightGBM, CatBoost)
  - **Stacking & Voting**: Meta-ensembles (Stacked Generalization, Voting Classifier/Regressor)
  - **AutoML**: Automated pipelines (Auto-sklearn, FLAML, TPOT, H2O AutoML, Optuna, Ray Tune)
  - **Neural Architecture Search**: Auto design (DARTS, ENAS, AutoKeras)
  - **Foundation Model Adaptation**: Parameter-efficient tuning (LoRA, QLoRA, Adapters, Prompt Tuning, In-context Learning)
