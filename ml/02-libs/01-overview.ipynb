## Data Manipulation Libraries

| Aspect                                              | NumPy                                                                                                 | Pandas                                                                                                                                      | Polars                                                                                                                 |
| --------------------------------------------------- | ----------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------- |
| Full Name                                           | Numerical Python                                                                                      | Panel Data (econometrics term)                                                                                                              | Process-Oriented Lazy Analytical Rust Series                                                                           |
| Primary Data Structure                              | `ndarray` (N-dimensional array)                                                                       | `Series` (1D), `DataFrame` (2D)                                                                                                             | `Series` (1D, homogeneous), `DataFrame` (2D, immutable)                                                                |
| Dimensionality                                      | Supports 0D to n-dimensional arrays                                                                   | Primarily 1D (Series) and 2D (DataFrame); higher via MultiIndex or xarray                                                                   | Primarily 1D (`Series`) and 2D (`DataFrame`); supports nested types like lists/structs                                 |
| Data Types Supported                                | Homogeneous (all elements same dtype)                                                                 | Heterogeneous (each column can have different dtype)                                                                                        | Homogeneous per column (`Series`); fixed types like `Int64`, `Float64`, ` String`, `Date`; nested support              |
| Memory Efficiency                                   | Extremely efficient (contiguous memory, fixed-type)                                                   | Less efficient (object overhead, pointers per cell)                                                                                         | Highly efficient (Arrow-based, columnar, no index overhead)                                                            |
| Performance (raw speed)                             | Fastest for numerical computations                                                                    | Slower than NumPy for pure math, but faster with vectorized operations on DataFrames                                                        | Significantly faster (Rust, multithreaded, lazy evaluation)                                                            |
| Missing Data Handling                               | No native NaN for integers (uses masked arrays or float NaNs)                                         | Excellent native support: `NaN`, `NA`, `NaT`, automatic alignment                                                                           | Unified `null` for all dtypes; `fill_null()` method                                                                    |
| Indexing                                            | Integer/location-based, boolean, fancy indexing                                                       | Label-based (`loc`), integer-based (`iloc`), boolean, MultiIndex, time-based                                                                | Positional only (no index/MultiIndex); integer-based                                                                   |
| Time Series Support                                 | Basic (requires manual work)                                                                          | Rich: `DatetimeIndex`, resampling, rolling windows, time zone support                                                                       | Strong: temporal types (`Date`, `Datetime`), window functions, group-by-time                                           |
| IO Tools                                            | Limited (`.npy`, `.npz`, basic text/CSV)                                                              | Supports: CSV, Excel, JSON, HDF5, Parquet, SQL, Stata, SAS, clipboard, HTML, etc.                                                           | CSV, Parquet, JSON, IPC, scanning (lazy), database connectors                                                          |
| Integration with Databases                          | None built-in                                                                                         | Direct SQL read/write (`read_sql`, `to_sql`)                                                                                                | Lazy scanning from SQL, PostgreSQL, MySQL                                                                              |
| Handling Labeled Data                               | Not designed for labels                                                                               | Core philosophy: everything is labeled                                                                                                      | No labels/index; focuses on columnar, query-based access                                                               |
| Use Case - When to Prefer                           | - Pure numerical computations<br>- Multidimensional data (images, tensors)<br>- High-performance math | - Tabular data analysis<br>- Data cleaning & preprocessing<br>- Exploratory data analysis (EDA)<br>- Working with real-world messy datasets | - Large-scale data processing<br/>- High-performance queries<br/>- Memory-constrained environments<br/>- ETL pipelines |
| Typical File Size for 1M rows x 10 cols (int/float) | ~80-100 MB (pure NumPy)                                                                               | ~200-400 MB (Pandas, depending on dtypes)                                                                                                   | ~50-100 MB (Arrow columnar, optimized dtypes)                                                                          |
| Relationship                                        | Pandas is built on top of NumPy                                                                       | Historically depended on NumPy arrays; moving toward pluggable backends (Arrow, etc.)                                                       | Built on Apache Arrow; Rust-native, Python/Rust bindings                                                               |
| Tip                                                 | Learn NumPy first to understand data structures and numerical operations                              | Leverage Pandas for data manipulation and analysis tasks                                                                                    | Use Polars for high-performance data processing and large datasets                                                     |

## Data Versioning

| Aspect                        | DVC (Data Version Control)                                                                                                                          | lakeFS                                                                                                                                                                    | Apache Iceberg                                                                                                                                                       |
| ----------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Overview**                  | Open-source Git extension for versioning data, models, and experiments in ML/DS projectsTracks metadata in Git while storing large files externally | Open-source Git-like version control system for data lakes on object storageEnables branching, merging, and atomic operations across any data format without copying data | Open-source table format for analytic datasetsProvides schema evolution, ACID transactions, and versioning via immutable snapshots for structured data in data lakes |
| **Primary Use Case**          | Reproducible ML pipelines, experiment tracking, and data/model versioning in development workflows                                                  | Scalable data lake management, ETL experimentation, collaboration, and rollback in production data lakes                                                                  | Reliable querying and management of large tabular data in lakehouses with time travel and schema flexibility                                                         |
| **Versioning Mechanism**      | Content-addressable metadata files (.dvc) with MD5 hashes stored in Git; full data files cached locally or remotely, deduplicated by content        | Git-style commits using metadata pointers (prolly trees) on object storage; zero-copy versioning with immutable objects                                                   | Immutable snapshots of table metadata; each write creates a new snapshot capturing the full table state                                                              |
| **Branching Support**         | No native branching; relies on Git branches for metadata, but data checkout required per branch, leading to copies                                  | Full Git-like branching: cheap, zero-copy branches for parallel experimentation; supports merge conflicts and diffs                                                       | Table-level branching and tagging via branch-aware catalogs (e.g., Nessie); not native to core format                                                                |
| **Merging Capabilities**      | No native data merging; manual resolution via Git for metadata, data conflicts handled by re-adding files                                           | Git-like merge with automatic conflict detection (e.g., uncommitted changes); supports three-way merges and hooks                                                         | Merge via snapshot management or catalog tools; row-level merges supported but no Git-style conflict resolution                                                      |
| **Time Travel / Rollback**    | Via Git history: checkout past commits to restore data versions; supports snapshots and restores                                                    | Full Git-like time travel: checkout any commit/ref; instant rollback across repositories without data movement                                                            | Native time travel: query by snapshot ID or timestamp; rollback by setting current snapshot to a previous one                                                        |
| **Concurrency Control**       | Git-based locking for metadata; no built-in ACID for data operations                                                                                | Atomic commits with optimistic locking; supports hooks for validation; no full ACID but ensures consistency via metadata                                                  | ACID transactions (atomic, consistent, isolated, durable) via snapshot isolation; handles concurrent writes safely                                                   |
| **Storage Backend**           | Local FS, cloud (S3, GCS, Azure Blob, Google Drive, HDFS, SSH/SFTP, HTTP); requires caching to local FS for access                                  | Object storage only (S3-compatible: AWS S3, GCS, Azure Blob, MinIO); no local FS support                                                                                  | File formats (Parquet, ORC, Avro) on object storage (S3, GCS, HDFS); metadata in catalogs (Hive, Glue, Nessie)                                                       |
| **Data Formats Supported**    | Any (files/directories: CSV, images, models, unstructured); agnostic but optimized for ML artifacts                                                 | Any (unstructured, semi-structured, structured like Iceberg/Delta/Hudi tables); format-agnostic                                                                           | Primarily structured tabular (Parquet, ORC, Avro); supports row-level updates/deletes via delete files                                                               |
| **Schema Evolution**          | Not applicable (file-based); schemas managed externally in code/pipelines                                                                           | Not native; depends on underlying formats (e.g., Iceberg integration for schema changes)                                                                                  | Native: add/drop/reorder/rename columns without rewriting data; promotes/demotes types safely                                                                        |
| **Scalability**               | Suitable for low-tens-of-thousands of files; performance degrades with millions due to local caching and hashing                                    | Handles billions of objects/petabyte-scale; low-overhead metadata (prolly trees); distributed access via S3 API                                                           | Petabyte-scale tables; hierarchical metadata for efficient scans; optimized for analytic queries on massive datasets                                                 |
| **Integration with Git**      | Tight: metadata files committed to Git; full Git workflow for reproducibility                                                                       | Loose: Git integration via hooks or external tools; focuses on data, not code                                                                                             | None native; can use Git for code, but data versioning via snapshots/catalogs                                                                                        |
| **Performance Optimizations** | Content deduplication, run caching for pipelines; auxiliary indices for diffs                                                                       | Zero-copy clones, pre-signed URLs for direct access; low-latency metadata ops (e.g., with S3 Express One Zone)                                                            | Hidden partitioning, file pruning, columnar stats; vectorized reads; compaction/rewrite for maintenance                                                              |
| **Limitations**               | Requires local checkout/caching (inefficient at scale); no ACID; relational DB features lacking; slower for very large datasets                     | Requires object storage; no built-in relational features; some advanced features (e.g., full Iceberg catalog) in enterprise                                               | Table-format only (not full repo management); branching needs external catalog; limited to structured data                                                           |
